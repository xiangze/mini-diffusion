```python
p=False
funcs=[]

with open("train.py") as fp:
    ls=fp.readlines()
    for l in ls:
        if(p):
            ll=l.replace(" ","").replace("),",")").replace("\n","").replace(",","@").replace("tdot(","")
            ll=ll.replace("_f","").replace("tsum","sum").replace("tmean","mean").replace("xd","dx").replace("xe","x")
            funcs.append(ll)            
        if("vs=[" in l):
            p=True
        elif("]" in l):
            p=False
            
for f in funcs:
    print(f)
```
